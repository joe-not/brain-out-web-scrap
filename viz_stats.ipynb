```python
kills_columns=[]
for i in df_positiv.columns:
    if i.startswith('kills-from-weapon-'):
        kills_columns.append(i)
```

```python
idx =[]
for i in kills_columns:
    idx.append(df_positiv[i].idxmax()+1)
    print('[tr]\n    [th]',i,'[/th]\n    [th]', df_positiv[i].idxmax()+1,'[/th]\n    [th]',df_positiv[i].max(),'[/th]\n[/tr]')
    #print(i, df_positiv[i].idxmax()+1, df_positiv[i].max())
```

```python
idx.sort()
len(idx)
```

```python
idx_set = set(idx)
achvm_count = dict()
for i in idx_set:
    achvm_count[i] = idx.count(i)
```

```python
achvm = pd.Series(achvm_count)
achvm.sort_values(inplace = True)
top_achvm = achvm[achvm>1]
```

```python
f, ax = plt.subplots(figsize=(15, 10))
sns.barplot(y=top_achvm.index,x=top_achvm.values,label = 'Achvm-count', orient = 'h',order =top_achvm.index)
ax.legend(ncol=2, loc="lower left", frameon=True)
ax.set(ylabel="ids",
       xlabel="Number of achievements")
sns.despine(left=True, bottom=True)
```

```python
kills_deaths = df_positiv.groupby('kills')['deaths','ids'].max()

kills_deaths = kills_deaths.tail(20)

kills_deaths.ids = kills_deaths.ids.astype(str)
kills_deaths = kills_deaths.sort_values("kills", ascending=False)
kills_deaths.reset_index(inplace = True)
```

```python
sns.set(style="whitegrid")

# Initialize the matplotlib figure
f, ax = plt.subplots(figsize=(15, 10))


# Plot the total crashes
sns.set_color_codes("pastel")
sns.barplot(x="kills", y="ids", data=kills_deaths,label = 'Kills',color="b",orient='h',order = kills_deaths.ids)

# Plot the crashes where alcohol was involved
sns.set_color_codes("muted")
sns.barplot(x="deaths", y="ids", data=kills_deaths,label = 'Deaths', color="r",orient='h',order = kills_deaths.ids)


# Add a legend and informative axis label
ax.legend(ncol=2, loc="lower left", frameon=True)
ax.set(xlim=(0, 310000), ylabel="ids",
       xlabel="Top 20 Kills - Deaths")
sns.despine(left=True, bottom=True)
```

```python
head_long = df_positiv.groupby('headshots')['ids','longshots'].max()
head_long = head_long.tail(20)

head_long.ids = head_long.ids.astype(str)
head_long = head_long.sort_values("headshots", ascending=False)
head_long.reset_index(inplace = True)
```

```python
sns.set(style="whitegrid")

# Initialize the matplotlib figure
f, ax = plt.subplots(figsize=(15, 10))

sns.set_color_codes("muted")
sns.barplot(x="headshots", y="ids", data=head_long,label = 'Headshots', color="b",orient='h',order = head_long.ids)

sns.set_color_codes("muted")
sns.barplot(x="longshots", y="ids", data=head_long,label = 'Longshots', color="g",orient='h',order = head_long.ids)

ax.legend(ncol=2, loc="lower left", frameon=True)
ax.set(xlim=(0, 200000), ylabel="ids",
       xlabel="Top 20 Head-Long shots")
sns.despine(left=True, bottom=True)
```

```python
#games-won - games-lost
winners = df_positiv.groupby('games-won')['games-lost','ids'].max()

winners = winners.tail(10)

winners.ids = winners.ids.astype(str)
winners = winners.sort_values("games-won", ascending=False)
winners.reset_index(inplace = True)
```

```python
games_win = winners.drop(columns = 'games-lost')
games_win['Win-lose'] = 'won'
games_win.rename(columns = {'games-won':'games-won-lose'},inplace = True)
```

```python
games_lose = winners.drop(columns = 'games-won')
games_lose['Win-lose'] = 'lose'
games_lose.rename(columns = {'games-lost':'games-won-lose'},inplace = True)
```

```python
games_winners = pd.concat([games_win,games_lose])
```

```python
sns.set(style="whitegrid")
g = sns.catplot(x="ids", y="games-won-lose", hue="Win-lose",order = games_win.ids, data=games_winners,
                height=8, kind="bar", palette="muted")
g.despine(left=True)
g.set_ylabels("Win-lose compare")
```
