```python
medal_columns =[]
for i in items.columns:
    if 'medal' in i:
        medal_columns.append(i)
```

```python
medal_columns.sort()
```

```python
medal_counts ={}
for i in medal_columns:
    if i == 'medal-steam':
        medal_counts[i] = items[i].value_counts().sum() - 41
    else:
        medal_counts[i] = items[i].value_counts().sum()
```

```python
medal_series = pd.Series(medal_counts)
```

```python
casual_medals = ['medal-1year', 'medal-2year', 'medal-f', 'medal-fighter',
                 'medal-master', 'medal-operation', 'medal-spectator', 'medal-wow']
unique_medals = ['medal-2ch', 'medal-embassador', 'medal-pikabu', 'medal-purple-heart',
                   'medal-star', 'medal-steam', 'medal-templar','medal-tube']
skin_medals = ['medal-skin-maker','medal-skin-maker-2','medal-skin-maker-3',
               'medal-skin-maker-4','medal-skin-maker-5']
tournament_medals = ['medal-tournament-1','medal-tournament-2','medal-tournament-3']
wiki_medals = ['medal-wiki','medal-wiki-2','medal-wiki-3']
```

```python
cas = medal_series[casual_medals].sort_values(ascending = False)
uni = medal_series[unique_medals].sort_values(ascending = False)
skin = medal_series[skin_medals].sort_values(ascending = False)
tourn = medal_series[tournament_medals].sort_values(ascending = False)
wiki = medal_series[wiki_medals].sort_values(ascending = False)
```

```python
grid = plt.GridSpec(3, 2, wspace=0.4, hspace=0.3)
sns.set_style("darkgrid")
```

```python
fig = plt.figure(figsize=(8, 9))
ax1 = fig.add_subplot(grid[0, 0])
ax1 = sns.barplot(y =uni.index, x=uni.values, 
                  palette=sns.color_palette("Paired"))
for p in ax1.patches:
    _x = p.get_x() + p.get_width() + float(2)
    _y = p.get_y() + p.get_height() - 0.15
    value = int(p.get_width())
    ax1.text(_x, _y, value, ha="left")
ax1.set(ylabel="Название медали",
       xlabel="Сколько игроков получило")
ax1.set_xticks(np.arange(0, 251, step=50))

ax2 = fig.add_subplot(grid[0, 1])
ax2 = sns.barplot(y =skin.index, x=skin.values, 
                  palette=sns.color_palette("Paired"))
for p in ax2.patches:
    _x = p.get_x() + p.get_width() + float(1)
    _y = p.get_y() + p.get_height() - 0.3
    value = int(p.get_width())
    ax2.text(_x, _y, value, ha="left")
ax2.set(xlabel="Сколько игроков получило")
ax2.set_xticks(np.arange(0, 100, step=25))

ax3 = fig.add_subplot(grid[1, 0])
ax3 = sns.barplot(y =tourn.index, x=tourn.values, 
                  palette=sns.color_palette("Paired"))
for p in ax3.patches:
    _x = p.get_x() + p.get_width() + float(1)
    _y = p.get_y() + p.get_height() - 0.3
    value = int(p.get_width())
    ax3.text(_x, _y, value, ha="left")
ax3.set(ylabel="Название медали",xlabel="Сколько игроков получило")

ax4 = fig.add_subplot(grid[1, 1])
ax4 = sns.barplot(y =wiki.index, x=wiki.values, 
                  palette=sns.color_palette("Paired"))
for p in ax4.patches:
    _x = p.get_x() + p.get_width() + float(0.5)
    _y = p.get_y() + p.get_height() - 0.3
    value = int(p.get_width())
    ax4.text(_x, _y, value, ha="left")
ax4.set(xlabel="Сколько игроков получило")
ax5 = fig.add_subplot(grid[2, :2])
ax5 = sns.barplot(y =cas.index, x=cas.values, palette=sns.color_palette("Paired"))
for p in ax5.patches:
    _x = p.get_x() + p.get_width() + float(100)
    _y = p.get_y() + p.get_height() - 0.3
    value = int(p.get_width())
    ax5.text(_x, _y, value, ha="left")
ax5.set(ylabel="Название медали",
       xlabel="Сколько игроков получило")
plt.subplots_adjust(left = -1, bottom = -0.5)
```
